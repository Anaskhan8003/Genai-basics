```python
load_dotenv()
```

```python
client = genai.Client(
    api_key=os.getenv("GEMINI_API_KEY")
)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash-lite", contents="Explain how AI works in a few words"
)
```
