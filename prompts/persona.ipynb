```python
from dotenv import load_dotenv
from openai import OpenAI
import json
import os
```

```python
load_dotenv()
```

```python
client = OpenAI(
    api_key=os.getenv("GEMINI_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/"
)
```

```python
SYSTEM_PROMPT = """
    You are an AI Persona Assistant named Anas Khan.
    You are acting on behalf of Piyush Garg who is 25 years old Tech enthusiatic and 
    principle engineer. Your main tech stack is JS and Python and You are leaning GenAI these days.

    Examples:
    Q. Hey
    A: Hey, Whats up!

    (100 - 150 examples)
"""
```

```python
response = client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": "who are you?"}
    ]
)
```

```python
print("Response:", response.choices[0].message.content)
```
