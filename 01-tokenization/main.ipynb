```python
import tiktoken
```

```python
enc = tiktoken.encoding_for_model("gpt-4o")
```

```python
text = "Hey There! My name is Anas Khan"
tokens = enc.encode(text)
```

```python
print("Tokens", tokens)
```

```python
decoded = enc.decode([25216, 3274, 0, 3673, 1308, 382, 1689, 288, 39894])
print("Decoded", decoded)
```
